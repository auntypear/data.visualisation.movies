```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')

popular_movies = ratings['movieId'].value_counts().head(10)
top_movies = movies[movies['movieId'].isin(popular_movies.index)]

plt.figure(figsize=(10, 6))
sns.barplot(x=popular_movies.values, y=top_movies['title'], color=sns.color_palette('rocket')[0])
plt.title('Top 10 Most Rated Movies')
plt.xlabel('Number of Ratings')
plt.ylabel('Movie Title')
plt.show()
```
