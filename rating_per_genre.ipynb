```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')

# Split the genres and explode the dataframe
movies['genres'] = movies['genres'].str.split('|')
exploded_movies = movies.explode('genres')

# Merge ratings with exploded movies dataframe
genre_ratings = ratings.merge(exploded_movies, on='movieId')

# Calculate the average rating for each genre
avg_genre_rating = genre_ratings.groupby('genres')['rating'].mean().sort_values()

# Plot the average rating by genre
avg_genre_rating.plot(kind='bar', figsize=(12, 6), color='pink')
plt.title('Average Rating by Genre')
plt.ylabel('Average Rating')
plt.xlabel('Genre')
plt.show()
```
